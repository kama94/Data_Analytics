```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display
pd.options.display.max_columns = None
```

```python
df = pd.read_csv('Project_clean_data_actual.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['MS Zoning_I (all)']
```

```python
df['SalePrice']
```

```python
X = df.drop('SalePrice',axis=1)
```

```python
X.head()
```

```python
#Create  and y 
X = df.drop('SalePrice',axis=1)
y = df['SalePrice']

from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_model = ElasticNet()
```

```python
param_grid = {'alpha':[0.1,1,5,10,50,100],'l1_ratio':[0.1,0.5,0.7,0.9,0.95,1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=base_elastic_model,param_grid=param_grid,scoring='neg_mean_squared_error',cv=140,verbose=2)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_params_
```

```python
grid_model.best_estimator_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
y_pred = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mean_squared_error(y_test,y_pred)
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,y_pred))
RMSE
```

```python
test_prob = y_test - y_pred
```

### 4 way to test model
